```python
import subprocess

def get_connected_gpus():
    try:
        result = subprocess.run(["nvidia-smi", "--list-gpus"], capture_output=True, text=True)
        if result.returncode == 0:
            gpu_list = result.stdout.strip().split('\n')
            return gpu_list
        else:
            return None
    except FileNotFoundError:
        print("nvidia-smi not found. Make sure NVIDIA GPU drivers are installed.")
        return None

if __name__ == "__main__":
    connected_gpus = get_connected_gpus()
    if connected_gpus is not None:
        for i, gpu in enumerate(connected_gpus):
            print(f"GPU {i}: {gpu}")
    else:
        print("Unable to retrieve GPU information.")
```

```python
import requests

url = "https://github.com/WongKinYiu/yolov9/releases/download/v0.1/yolov9-s.pt"
filename = "yolov9-s.pt"

response = requests.get(url)
if response.status_code == 200:
    with open(filename, 'wb') as f:
        f.write(response.content)
    print("Download successful")
else:
    print("Failed to download:", response.status_code)
```
